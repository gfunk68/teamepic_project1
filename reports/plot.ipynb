```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Resources/almost_complete.csv')
df=df.drop(['Unnamed: 0'],axis=1)
df
```

```python
df.columns
```

```python
df.rename(columns = {'Life Ladder':'Happiness Score'}, inplace = True) 
```

```python
df1=df[df['Year']==2018]
df_u=df1.sort_values(by='Happiness Score',ascending=False)
df_u
```

```python
df_use=df_u.reset_index()
df_use
```

```python
happiest_rank = df_use[0:20]
x= happiest_rank[['Country','Happiness Score','Log GDP per capita','Social support','Healthy life expectancy at birth','Freedom to make life choices','Generosity','Perceptions of corruption','Positive affect','Negative affect','Confidence in national government']]
y= x.set_index('Country')
y
```

```python
plt.figure(figsize=(20,10))
y.plot(kind='bar',stacked=True)
plt.title('Countries with Top 20 Happiness Scores in 2018')
plt.ylabel('Happiness Score')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
# plt.savefig("./Images/Top_20.png")
```

```python
happiest_lowrank = df_use[115:135]
x1= happiest_lowrank[['Country','Happiness Score','Log GDP per capita','Social support','Healthy life expectancy at birth','Freedom to make life choices','Generosity','Perceptions of corruption','Positive affect','Negative affect','Confidence in national government']]
y1= x1.set_index('Country')
```

```python
plt.figure(figsize=(20,10))
y1.plot(kind='bar',stacked=True)
plt.title('Countries with Lowest 20 Happiness Scores in 2018')
plt.ylabel('Happiness Score')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
# plt.savefig("./Images/Low_20.png")
```

```python
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
data_select = df19
econ_size  = data_select['GDP per capita']*30
social_color = data_select['Social support']
data = [
    {
        'y': data_select['Freedom to make life choices'],
        'x': data_select['Overall rank'],
        'mode': 'markers',
        'marker': {
            'color': social_color,
            'size': econ_size,
            'showscale': True
        },
        "text" :  data_select.Country    
    }
]
layout = dict(title='Freedom vs Happiness Rank: Size=Economy, Color=Family',
              xaxis=dict(title='Happiness Rank'),
              yaxis=dict(title='Freedom'))
fig = go.Figure(data=data,layout=layout)
iplot(fig)
```
