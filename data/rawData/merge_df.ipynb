```python
import os
import pandas as pd
from api_call import worldbank,consumer_electronics
import sys
```

```python
#read both csvs
#if api fails, rely on backup CSV
try:
    worldbank_df = worldbank()
except:
    print("Oops!",sys.exc_info()[0],"occured, using CSV Instead")
    worldbank_df = pd.read_csv(os.path.join('Resources','worldbank.csv'))
try:
    CountryRevenueInfo = consumer_electronics()
except:
    print("Oops!",sys.exc_info()[0],"occured, using CSV Instead")
    CountryRevenueInfo = pd.read_csv(os.path.join('Resources','Consumer_Electronics_Sales.csv'))
    
worldhappiness_df = pd.read_csv(os.path.join('Resources', 'worldhappiness_prepped.csv'))

#join both data sources on Country Name, drop all rows that do not match

world_df = pd.merge(worldbank_df,worldhappiness_df, on='Country Code', how='outer')
world_df = pd.merge(world_df,CountryRevenueInfo, on='Country Code', how='outer')
```
